```python
#2014, 2013년 데이터의 형식을 동일하게 맞춰준다.
#2014
# pivot_table을 사용하여 데이터 정렬
# market_data2014 = market_data2014.pivot_table(index='품목별', columns='항목', values='2014 년', aggfunc='first')
# # 컬럼명 수정
# market_data2014.columns.name = None
# market_data2014.head(5)

market_data2013 = market_data2013.pivot_table(index='품목별', columns='항목', values='2013 년', aggfunc='first')
# 컬럼명 수정
market_data2013.columns.name = None
market_data2013.head(5)

```

```python
#컬럼명 변경 (2019~16, 2006~2003 : 컬럼명 20XX.2 -> 출하액[백만원])
market_data2019.rename(columns={'2019.2':'출하금액[백만원]'}, inplace=True)
market_data2018.rename(columns={'2018.2':'출하금액[백만원]'}, inplace=True)
market_data2017.rename(columns={'2017.2':'출하금액[백만원]'}, inplace=True)
market_data2016.rename(columns={'2016.2':'출하금액[백만원]'}, inplace=True)
market_data2006.rename(columns={'2006.2':'출하금액[백만원]'}, inplace=True)
market_data2004.rename(columns={'2004.2':'출하금액[백만원]'}, inplace=True)
market_data2003.rename(columns={'2003.2':'출하금액[백만원]'}, inplace=True)
```

```python
# market_data2012 ~ 07은 컬럼이 없다. 컬럼 설정을 해줘야한다. 
```

```python
market_data2019_sub = market_data2019[["품목별","출하금액[백만원]"]]
market_data2018_sub = market_data2018[["품목별","출하금액[백만원]"]]
market_data2017_sub = market_data2017[["품목별","출하금액[백만원]"]]
market_data2016_sub = market_data2016[["품목별","출하금액[백만원]"]]

market_data2014_sub = market_data2014[["품목별","출하금액[백만원]"]]
market_data2013_sub = market_data2013[["품목별","출하금액[백만원]"]]
#2012~07

market_data2006_sub = market_data2006[["품목별","출하금액[백만원]"]]
market_data2004_sub = market_data2004[["품목별","출하금액[백만원]"]]
market_data2003_sub = market_data2003[["품목별","출하금액[백만원]"]]
```

```python
merged_market_df_16_19 = pd.merge(market_data2016_sub, market_data2017_sub, on= "품목별", how='outer')
merged_market_df_16_19 = pd.merge(merged_market_df_16_19, market_data2018_sub, on= "품목별", how='outer')
merged_market_df_16_19 = pd.merge(merged_market_df_16_19, market_data2019_sub, on= "품목별", how='outer')
```

```python
dfrow = merged_market_df_16_19[merged_market_df_16_19.isin(['김치'])]
dfrow
```

>시각화

```python
specific_item = '김치'
df_specific_item = merged_market_df_16_19[merged_market_df_16_19['품목별'] == specific_item]
# Melt the DataFrame to convert columns into rows for easier plotting
df_melted = df_specific_item.melt(id_vars='품목별', var_name='Year', value_name='Production Amount')
df_melted['Year'] = df_melted['Year'].str.split('.').str[0].astype(int)

df_melted
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Production Amount', data = df_melted, marker='o', markersize=8)
plt.title("Production Amount by Item (2016-2019)")
plt.xlabel("Year")
plt.ylabel("Production Amount")
plt.grid(True)
plt.legend(title='Item', loc='best', bbox_to_anchor=(1, 1))

# Customize the y-axis formatting and ticks
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))  # Format with commas and no decimal places
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(base=100000))  # Set the tick interval to 1000
plt.gca().invert_yaxis()

plt.show()
```
