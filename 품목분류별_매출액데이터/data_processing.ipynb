```python
df = pd.read_csv('매출액_2019.csv', engine='python',encoding='utf-8')
df2 = pd.read_csv('매출액_2018.csv', engine='python',encoding='utf-8')
df
```

```python
df_con = pd.concat((df,df2))
df_con
# df_con.to_csv(output_file, index=False)
```

```python
# 품목을 기준으로 그룹화하고, cumcount 함수를 사용하여 중복되는 값의 인덱스 생성
df_con['인덱스'] = df_con.groupby('품목별').cumcount() + 1

# pivot()을 사용하여 데이터 변경
df_pivot = df_con.pivot(index='품목별', columns='인덱스', values='출하액 (백만원)')

# 열 이름 변경
df_pivot.columns = [f'값{col}' for col in df_pivot.columns]

# 인덱스 재설정
df_pivot.reset_index(inplace=True)
```

```python
df_pivot
df_pivot.to_csv('wow.csv',index=False)
# 성공 하였다. 
#이때 concat을 2019, 2018 순으로 하여 pivot을 진행하여서, 컬럼의 값이 품목별 2019 2018 -> 먼저 온 순서대로 된다. 
```

```python
# 마지막 확인으로 3개의 파일을 동시에 진행하면 어떻게 될까?
df1 = pd.read_csv('매출액_2019.csv', engine='python',encoding='utf-8')
df2 = pd.read_csv('매출액_2018.csv', engine='python',encoding='utf-8')
df3 = pd.read_csv('매출액_2017.csv', engine='python',encoding='utf-8')

df_con = pd.concat((df3,df2,df1))
# 품목을 기준으로 그룹화하고, cumcount 함수를 사용하여 중복되는 값의 인덱스 생성
df_con['인덱스'] = df_con.groupby('품목별').cumcount() + 1

# pivot()을 사용하여 데이터 변경
df_pivot = df_con.pivot(index='품목별', columns='인덱스', values='출하액 (백만원)')

# 열 이름 변경
df_pivot.columns = [f'값{col}' for col in df_pivot.columns]

# 인덱스 재설정
df_pivot.reset_index(inplace=True)

df_pivot
df_pivot.to_csv('wow_2.csv',index=False)
```
